```python
import numpy as np
import pandas as pd
```

```python
training_data = pd.read_csv('credit_card_default_train.csv',index_col=0)
training_x = training_data.drop('NEXT_MONTH_DEFAULT',1)
training_y = training_data['NEXT_MONTH_DEFAULT']
test_data = pd.read_csv('credit_card_default_test.csv',index_col=0)
```

```python
data_comp = training_x.append(test_data)
```

```python
def intoIntegers(Col_name,titles,dataset):
    for i in range(len(dataset)):
        dataset[Col_name][i] = titles.index(dataset[Col_name][i])
```

```python
intoIntegers("Gender",['M','F'],data_comp)
intoIntegers("EDUCATION_STATUS",['Graduate','High School','Other'],data_comp)
intoIntegers("AGE",['Less than 30','31-45','46-65','More than 65'],data_comp)
intoIntegers("MARITAL_STATUS",['Single','Other'],data_comp)
```

```python
data_comp.head()
```

```python
for i in range(len(data_comp)):
    item = data_comp['Balance_Limit_V1'][i]
    if item[-1] == 'M':
        data_comp['Balance_Limit_V1'][i] = float(item[:-1])* (10**6)
    elif item[-1] == 'K':
        data_comp['Balance_Limit_V1'][i] = float(item[:-1])* (10**3)
```

```python
data_comp.head()
```

```python
data_comp = data_comp.astype('int64')
```

```python
training_x_modified = data_comp.iloc[:len(training_x)]
test_data_modified = data_comp.iloc[len(training_x):]
training_x_modified["NEXT_MONTH_DEFAULT"] = training_y
```

```python
training_x_modified.to_csv("training_data_modified.csv")
test_data_modified.to_csv("test_data_modified.csv")
```
